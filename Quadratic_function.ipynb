```python
import pandas as pd
import numpy as np
```

```python
numbers = []
for i in range(0,1000):
    numbers.append(i+1)
```

```python
numbers_df = pd.DataFrame(numbers, columns=['x'])
```

```python
numbers_df['y_sqrt'] = np.sqrt(numbers_df['x'])
numbers_df['y_linear'] = numbers_df['x']
```

```python
numbers_df.plot()
```
